```python
import numpy as np
import pandas as pd

df1 = pd.read_csv("../data/student-mat.csv")
df2 = pd.read_csv("../data/student-por.csv")
```

Nos deux DataFrames ont la même structure, donc on va procéder à leur concaténation

Mais avant la concaténation, nous allons créer une nouvelle colonne pour nos Dataframe afin
de bien préciser de quelle matière il s'agit.

```python
df1["matiere"] = "Maths"
df2["matiere"] = "Portugais"
```

```python
data = pd.concat([df1, df2])
data.head()
```

À présent, on exporte la DataFrame obtenue

```python
data.to_csv("../data/student_merged.csv", index=False)
```
