# Analyse Statistique et Manipulation

## Analyse Univariée

### Étude des variables qualitatives

Cette étude consistera principalement à comprendre la distribution de nos variables au sein de notre échantillon pour en tirer des conclusions pertinentes.


````python
data.head()
````

````python
liste_variables = ["school", "sex", "address", "taille_famille", "niveau_edu_mere", "niveau_edu_pere", "emploi_mere", "emploi_pere", "raison_choix_ecole", "tuteur", "matiere"]
````

````python
def add_labels(k, y):
    for i in range(len(k)):
        plt.text(i, y.iloc[i], y.iloc[i], horizontalalignment="center", verticalalignment="bottom")
for x in liste_variables:
    print("--"*50)

    table = pd.DataFrame(data[x].value_counts())
    table["frequency"] = (table["count"] / table["count"].sum()).round(3)
    display(table)

    print("Variable ", x)
    print("Mode ", data[x].mode()[0])

    table["frequency"].plot(kind="pie", autopct="%.1f%%")
    plt.show()

    add_labels(table.index, table["count"])
    table["count"].plot(kind="bar")
    plt.show()
````

````python
liste_variables = ["qualite_relation_fam", "freetime", "sorties_amis", "conso_alcool_workday", "conso_alcool_weekend", "health"]
````

````python
for x in liste_variables:
    print("--"*50)

    table = pd.DataFrame(data[x].value_counts())
    table["frequency"] = (table["count"] / table["count"].sum()).round(3)
    table.sort_values(x, inplace=True)
    display(table)

    print("Variable ", x)
    print("Mode ", data[x].mode()[0])

    table["frequency"].plot(kind="pie", autopct="%.1f%%")
    plt.show()

    add_labels(table.index, table["count"])

    table["count"].plot(kind="bar")
    plt.show()
````

À partir des résultats obtenus de cette analyse, on ne retiendra que le mode de chacune de nos variables.
